**컨볼루션 후 특징 맵의 개수가 증가**하면 Flatten 이후 입력으로 사용할 데이터의 컬럼수가 증가하게 되고 이는 **곧 컴퓨터가 찾아야할 가중치의 증가**를 의미한다.

**pooling**은 flatten이후 컴퓨터가 찾아야할 가중치를 작게 유지하기 위해 입력으로 사용할 컬럼수를 조정하는 것이 목적이다.

convolution으로 생성된 feature map에 activation function을 적용한 map을 **Activation map**이라고 한다. 이러한 Activation map의 크기를 줄이거나 강조하는 방법으로 pooling이 사용되는데  처리하는 방법으로는 **Max pooling, Average pooling, Min pooling**이 있다.

**Max pooling**은 Activation map을 pooling 크기로 쪼개서 해당 map안에 **가장 큰 값**을 가져오는 것을 말하며, 이 때, **stride값은 pooling 크기와 같도록 설정**한다.

Average pooling, Min pooling은 가장 큰 값이 아닌 **평균값** ,**가장 작은 값**을 가져오는 방식을 말한다.

```python
import tensorflow as tf
import pandas as pd

# 데이터 로드
(독립, 종속), _ = tf.keras.datasets.mnist.load_data()
독립 = 독립.reshape(60000,28,28,1) # 흑백이미지도 3차원 형태로 변환해 준다고 생각
종속 = pd.get_dummies(종속)
print(독립.shape, 종속.shape)
```

```python
# 모델 생성
X = tf.keras.layers.Input(shape=[28,28,1])

H = tf.keras.layers.Conv2D(3, kernel_size=5, activation='swish')(X)
H = tf.keras.layers.MaxPool2D()(H)
H = tf.keras.layers.Conv2D(6, kernel_size=5, activation='swish')(H)
H = tf.keras.layers.MaxPool2D()(H)

H = tf.keras.layers.Flatten()(H) # 표로 만든다 
H = tf.keras.layers.Dense(84, activation='swish')(H)
Y = tf.keras.layers.Dense(10, activation='softmax')(H)
model = tf.keras.models.Model(X,Y)
model.compile(loss='categorical_crossentropy', metrics='accuracy')
```

```python
# 모델 확인
model.summary()
```

```python
 # 모델 학습
 model.fit(독립, 종속, epochs=10)
```
