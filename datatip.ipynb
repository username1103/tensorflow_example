# 데이터를 범주형으로 변경



*   원핫 인코딩시 데이터가 category나 object타입이어야 pd.get_dummies(data)로 원핫인코딩가능



```python
import pandas as pd
```

```python
path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris2.csv'
data = pd.read_csv(path)

data.head()
```

```python
# 원 핫 인코딩
encoding_data = pd.get_dummies(data)
encoding_data.head()


# pandas는 숫자로 데이터가 되어있으면 숫자로인식하므로 범주형 데이터로 변환이 필요.
```

```python
print(data.dtypes)

# 아래에 타입이 category이거나 object일 때만 원핫인코딩을 하게 되기 때문에 변경해줘야함.
```

```python
# 품종 타입을 범주형으로 바꾸어 준다.
data['품종'] = data['품종'].astype('category')
print(data.dtypes)
```

```python
# 원 핫 인코딩
encoding_data = pd.get_dummies(data)
encoding_data.head()

# 원핫 인코딩된 결과를 얻을 수 있음.
```

# NA값 체크하기

```python
# NA값을 체크해보기
data.isna().sum()
```

```python
data.tail()
```

```python
# NA값에 꽃잎폭 평균값을 넣어주는 방법
mean = data['꽃잎폭'].mean()
data['꽃잎폭'] = data['꽃잎폭'].fillna(mean)
data.tail()
```
