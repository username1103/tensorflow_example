# 히든레이어를 사용한 보스턴 집값 예측

```python
import tensorflow as tf
import pandas as pd

# 데이터 준비하기
path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
data = pd.read_csv(path)

# print(data.columns) # 컬럼 내용 확인
independent = data[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax','ptratio', 'b', 'lstat']]
dependent = data[['medv']]
# print(independent.shape, dependent.shape) # 잘 나누었는지 확인

# 모델 생성
X = tf.keras.layers.Input(shape=[13])
H = tf.keras.layers.Dense(10, activation='swish')(X)
Y = tf.keras.layers.Dense(1)(H)
model = tf.keras.models.Model(X,Y)
model.compile(loss='mse')
```

```python
# 모델을 확인
model.summary()
```

```python
# 데이터로 모델을 학습
model.fit(independent,dependent,epochs=1000 ,verbose=0)
model.fit(independent,dependent,epochs=10)
```

```python
# 모델을 이용
print(model.predict(independent[0:5]))

# 비교
dependent[0:5]
```

```python
# 가중치 확인
weights = model.get_weights()


# 학습된 모델
print('medv = ',  end='')
for i in range(len(weights[0])):
  print(weights[0][i][0], end=' * x' + str(i + 1) + ' + ') 

# 편향
print(weights[1][0])
```

# 히든레이어를 사용한 아이리스 품종 구별

```python
import tensorflow as tf
import pandas as pd

path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
data = pd.read_csv(path)

# 원핫 인코딩
data = pd.get_dummies(data)

independent = data[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
dependent = data[['품종_setosa', '품종_versicolor','품종_virginica']]

# 모델 생성
X = tf.keras.layers.Input(shape=[4])
H = tf.keras.layers.Dense(5, activation='swish')(X)
Y = tf.keras.layers.Dense(3, activation='softmax')(H)
model = tf.keras.models.Model(X,Y)
model.compile(loss="categorical_crossentropy", metrics="accuracy")
```

```python
# 모델을 학습시킴
model.fit(independent, dependent, epochs=100)
```

```python
# 모델을 이용
print(model.predict(independent[-5:]))
type = ['setosa', 'versicolor','virginica'];
for result in model.predict(independent[-5:]):
  print(type[list(result).index(max(result))], end=' ')

# 비교
dependent[-5:]
```

```python
# 학습한 가중치
model.get_weights()

# list에서 같은 index에 있는 수들로 하나에 식을 만듬
```
